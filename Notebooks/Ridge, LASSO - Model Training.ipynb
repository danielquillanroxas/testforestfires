## **Algerian Forest Fires Dataset**

Dataset Information:

This dataset inclues 244 instances that regroup a data of two regions of Algeria, namely Bejaia region located in the northeast of Algeria and the Sidi Bel abbes region located in the northwest of Algeria. (122 instances for each region)

The period is between June 2012 and September 2012. The dataset includes 11 attributes and 1 output attribute (class). The 244 instances have been cclassified into **fire** (138 classes) and **not fire**(106 classes) classes.

Attribute Information:

    - Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012) Weather data observations
    - Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
    - RH : Relative Humidity in %: 21 to 90
    - Ws :Wind speed in km/h: 6 to 29
    - Rain: total day in mm: 0 to 16.8 FWI Components
    - Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
    - Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
    - Drought Code (DC) index from the FWI system: 7 to 220.4
    - Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
    - Buildup Index (BUI) index from the FWI system: 1.1 to 68
    - Fire Weather Index (FWI) Index): 0 to 31.1
    - Classes: two classes, namely Fire and not Fire

## Data Cleaning

```python
# missing values
dataset[dataset.isnull().any(axis=1)]
```

The dataset is converted into two sets on Region from 122nd index, we can make a column based on the Region

1 : 'Bejaia Region'
2 : 'Sidi-Bel Abbes Region Dataset'

To solve this, we have to add new column with region

```python
dataset.loc[:122, "Region"]=0
dataset.loc[122:, "Region"]=1
```

```python
dataset[['Region']] = dataset[['Region']].astype(int) 
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
# remove the null values (in row 122 and 167)
df = dataset.dropna().reset_index(drop=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]]
```

```python
df = df.drop(index=122).reset_index(drop=True)
```

```python
df.iloc[[122]]
```

```python
df.shape
```

```python
df.columns
```

```python
# fix spaces in columns
df.columns = df.columns.str.replace(' ', '')
```

```python
df.columns
```

```python
df.info()
```

### Change the columns from integer to data type where necessary

```python
df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']] = df[['month', 'day', 'year', 'Temperature', 'RH', 'Ws']].astype(int)
```

```python
df.info()
```

### Change the other columns to float data type

```python
objects = [features for features in df.columns if df[features].dtypes=='O']
```

```python
for i in objects:
    if i != 'Classes':
        df[i] = df[i].astype(float)
```

```python
df.info()
```

```python
# Save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv', index=False)
```
